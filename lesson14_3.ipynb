```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scroes_df = pd.DataFrame(scroes,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
scroes_df = pd.DataFrame(scroes,
                         columns=['國文','英文','數學','地理','歷史'],
                         index=range(1,51))
scroes_df
sum_values = scroes_df.sum(axis=1)
mean_values = scroes_df.mean(axis=1)
scroes_df['總分']=sum_values
scroes_df['平均']=mean_values
scroes_df
```

```python
scroes_df #上色
scroes_df.style\
    .format(precision=2)\
    .highlight_between(left=50,right=59,props='color:white; background-color:#8E354A;')\
    .highlight_max(axis=1,subset=scroes_df.columns[:5],props='color:white; background-color:#616138;')\
    .to_excel('小學.xlsx',sheet_name='3年1班')
```
