```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scroes_df = pd.DataFrame(scroes,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
scroes_df
```

```python
#'國文','英文','數學','地理','歷史'
scroes_df['總分']=scroes_df['國文']+scroes_df['英文']+scroes_df['數學']+scroes_df['歷史']+scroes_df['地理']
scroes_df['平均']=scroes_df['總分']/5
scroes_df
#算總分
```
