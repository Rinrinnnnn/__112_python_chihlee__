```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scroes_df = pd.DataFrame(scroes,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
scroes_df = pd.DataFrame(scroes,
                         columns=['國文','英文','數學','地理','歷史'],
                         index=range(1,51))
scroes_df
sum_values = scroes_df.sum(axis=1)
mean_values = scroes_df.mean(axis=1)
scroes_df['總分']=sum_values
scroes_df['平均']=mean_values
scroes_df
```

```python
scroes_df.sort_values(by='總分',axis=0) #按照平均分數高寡排序
```

```python
scroes_df.sort_values(by='總分',ascending=False).iloc[:5] 
#按照平均分數高寡排序
#取最後五名
```

```python
#使用subscript[]過濾資料
scroes_df[scroes_df['平均']<80]
```
